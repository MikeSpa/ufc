```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("fighters.csv", index_col=0)
print(df)
```

Let's change those "--" into proper null values

```python
df = df.replace({'--': np.nan})
print(df)
```

```python
print(df.isnull().sum())
```

Don't know how there is that many fighter with missing weight but we will removed them.

```python
df = df[df['Weight'].notna()]
print(df.isnull().sum())

print(f"New size: {df.shape[0]}")
```

Let's get rid of those ridiculous american unit. First the weight

```python
def pounds_to_kilos(s):
    w = s.split(" ", 1)[0]
    w = float(w) * 0.4535924
    return round(w, 1)

pounds_to_kilos(df.iloc[0,4])

df['Weight'] = df['Weight'].apply(lambda x: pounds_to_kilos(x))
df
```

```python
def feet_inches_to_cm(s):
    if type(s) == float and np.isnan(s): 
        return np.nan
    f = s.split("'", 1)[0]
    # print(f)
    i = (s.split(" "))[1].split("\"")[0]
    # print(i)
    h = float(f) * 30.48 + float(i) * 2.54
    return int(h)

# feet_inches_to_cm(df.iloc[0,3])

df['Height'] = df['Height'].apply(lambda x: feet_inches_to_cm(x))
df
```

```python
def inches_to_cm(s):
    if type(s) == float and np.isnan(s): 
        return np.nan
    i = s.split("\"", 1)[0]
    r = float(i) * 2.54
    return int(r)

# feet_inches_to_cm(df.iloc[0,3])

df['Reach'] = df['Reach'].apply(lambda x: inches_to_cm(x))
df
```
